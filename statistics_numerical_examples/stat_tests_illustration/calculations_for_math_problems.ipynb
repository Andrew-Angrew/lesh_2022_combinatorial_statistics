## Задача про шулера

```python
p_honest = 0.5
p_sharper = 0.9
sharper_prior = 0.1
honest_prior = 1 - sharper_prior

def posterior_sharper_prob(n_games, n_wins):
    n_losses = n_games - n_wins
    honest_likelihood = binom(n_games, n_wins) * (p_honest ** n_wins) * ((1 - p_honest) ** n_losses)
    sharper_likelihood = binom(n_games, n_wins) * (p_sharper ** n_wins) * ((1 - p_sharper) ** n_losses)
    p_observation = honest_prior * honest_likelihood + sharper_prior * sharper_likelihood
    return sharper_prior * sharper_likelihood / p_observation

def plot_posterior_sharper_prob(n_games):
    n_wins_range = list(range(n_games + 1))
    probs = [posterior_sharper_prob(n_games, n_wins) for n_wins in n_wins_range]
    plt.figure(figsize=(10, 5))
    plt.plot(n_wins_range, probs)
    plt.xlabel("Количество побед")
    plt.ylabel("Вероятность шулера")
    plt.show()
```

```python
plot_posterior_sharper_prob(10)
```

```python
plot_posterior_sharper_prob(20)
```
